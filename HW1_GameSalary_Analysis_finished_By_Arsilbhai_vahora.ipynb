```python
# 14)
# A) Show all Titles name with no duplicate,       
# B) Display the title names as flowing format    

#res = gameSalary['Title'].unique()
res = gameSalary.Title.unique()
print(res)

print('\nTitles')
print('----------')
for name in res:
    print(name)
```

```python
# 15) Create a bar chart to compare average Salary by gender. 
# .groupby('')      
# .mean()      
# plot    

avg_salary = gameSalary.groupby('Gender')['Salary'].mean()

avg_salary.plot(kind='bar')

plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.title('Average Salary by Gender')

plt.show()
```

```python
# 16)+17) 
# Create a bar chart to compare average Salary by Department. 
# .groupby('')      
# .mean()      
# plot    
avg_salary = gameSalary.groupby('Department')['Salary'].mean()

avg_salary.plot(kind='bar')

plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.title('Average Salary by Department')

plt.show()



```

```python
# 17) + 18) Show the number of employees for each performance level 
# and create pie chart (and show percentage)
counts = gameSalary['Performance'].value_counts()
print(counts)
counts.plot(kind='pie', autopct='%1.1f%%')

plt.title('Employees by Performance Level')
plt.ylabel('')  # removes extra label

plt.show()


```

```python
# 19) + 20) 
# Show the number of employee for each department,
# and sorted by the values from hight to low
# And plot the result as bar chart

#  .groupby('Department') or df.Department.value_counts() 
#  .sort_values(ascending=False)  
#  .plot 

dept_counts = gameSalary['Department'].value_counts().sort_values(ascending=False)

print(dept_counts)

dept_counts.plot(kind='bar')

plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Employees per Department')

plt.show()

```
